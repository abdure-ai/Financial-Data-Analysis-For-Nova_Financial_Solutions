```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
data = pd.read_csv(r'C:\Users\user\Projects\Financial-Data-Analysis-For-Nova_Financial_Solutions\Data\raw_analyst_ratings.csv')
```

```python
data.head(20)
```

```python
data['date'] = pd.to_datetime(data['date'], errors='coerce') 
```

```python
data.head()
```

```python
data.head()
```

```python
# 1. Descriptive Statistics: Headline Length
# ------------------------------
data['headline_length'] = data['headline'].str.len()  # Calculate length of each headline

# Basic statistics
headline_stats = data['headline_length'].describe()
print("\nHeadline Length Statistics:")
print(headline_stats)
```

```python
# Plot distribution of headline lengths
plt.figure(figsize=(10, 6))
sns.histplot(data['headline_length'], kde=True, bins=30, color='blue')
plt.title('Distribution of Headline Lengths')
plt.xlabel('Headline Length')
plt.ylabel('Frequency')
plt.show()
```

```python
# ------------------------------
# 2. Articles Per Publisher
# ------------------------------
publisher_counts = data['publisher'].value_counts()  # Count articles by publisher

# Display the counts
print("\nNumber of Articles Per Publisher:")
print(publisher_counts)

# Plot the top publishers
plt.figure(figsize=(12, 6))
publisher_counts.head(10).plot(kind='bar', color='green')  # Top 10 publishers
plt.title('Top 10 Most Active Publishers')
plt.xlabel('Publisher')
plt.ylabel('Number of Articles')
plt.show()
```

```python
# ------------------------------
# 3. Publication Date Analysis
# ------------------------------
# Group by date and count the number of articles
daily_articles = data.groupby(data['date'].dt.date).size()

# Display daily article trends
print("\nDaily Article Trends:")
print(daily_articles.head())

```

```python
# Plot the publication trends over time
plt.figure(figsize=(14, 7))
daily_articles.plot(color='purple')
plt.title('Articles Published Over Time')
plt.xlabel('Date')
plt.ylabel('Number of Articles')
plt.show()
```

```python
# Optional: Analyze spikes or trends by days of the week
data['day_of_week'] = data['date'].dt.day_name()
day_counts = data['day_of_week'].value_counts()

# Plot publication trends by day of the week
plt.figure(figsize=(10, 6))
day_counts.plot(kind='bar', color='orange')
plt.title('Articles Published by Day of the Week')
plt.xlabel('Day of the Week')
plt.ylabel('Number of Articles')
plt.show()
```

```python
from nltk.sentiment.vader import SentimentIntensityAnalyzer
sia = SentimentIntensityAnalyzer()
```

```python
data['sentiment_score'] = data['headline'].apply(lambda x: sia.polarity_scores(x)['compound'])
```

```python
data['sentiment'] = data['sentiment_score'].apply(lambda x: 'Positive' if x > 0.1 else ('Negative' if x < -0.1 else 'Neutral'))
```
